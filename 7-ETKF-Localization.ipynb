## localized version

```python
def integers(a, b):  # for periodic condition
    return [i for i in range(a, b+1)]


def inverse_gaussian_function(x, y, sigma):
    return np.exp((x**2 + y**2) / (2 * sigma**2))

def apply_gaussian_filter(matrix, sigma):
    rows, cols = matrix.shape
    assert rows == cols, "Matrix must be square"

    size = rows
    center = size // 2

    new_matrix = np.zeros_like(matrix)
    for i in range(size):
        for j in range(size):
            if i == center and j == center:
                new_matrix[i, j] = matrix[i, j]
            else:
                x = j - center
                y = i - center
                g_val = inverse_gaussian_function(x, y, sigma)
                new_matrix[i, j] = g_val * matrix[i, j]
                
    return new_matrix


def member2Z(ensemble, ensemble_mean):   # calculate Z from ensemble and ensemble mean
    ensemble_perturbed = []
    for ii in range(m):
        ensemble_perturbed.append(ensemble[ii]-ensemble_mean)
    Z = np.zeros((n,m))
    for ii in range(m):
        Z[:,ii] = ensemble_perturbed[ii] /np.sqrt(m-1)
    return Z

def Z2member(Z,ensemble_mean):
    ensemble = []
    for ii in range(m):
        ensemble.append(np.real(ensemble_mean + np.sqrt(m-1)*Z[:,ii]))
    return ensemble

def lorenz96_forecast(ensemble,ensemble_mean_history,x_history,x_mean_history):
    
    for jj in range(analysis_intv):
        for ii in range(m):
            ensemble[ii] = lorenz96_step(ensemble[ii])
        xxx = np.array(ensemble); xxx = xxx[:,0]; x_history.append(xxx); x_mean_history.append(np.mean(xxx))
        ensemble_mean_history.append(np.mean(ensemble,axis = 0))
    ensemble_mean = np.mean(ensemble,axis = 0)
    Z = member2Z(ensemble,ensemble_mean)

    return ensemble, ensemble_mean, ensemble_mean_history, x_history,x_mean_history, Z
    
def lorenz96_analysis(ensemble, ensemble_mean, Z, H, y, R):
     
    ensemble_mean_new = ensemble_mean
    Z_new = Z
    for idx in range(n): # n = 40, the Lorenz96 size
        idx_range = np.array(integers(idx-cutoff,idx+cutoff))%n  # periodic condition
        # Step 1: get local y, Y, R, Z, K, T
        H2 = H[idx_range,:]
        y2 = H2@y
        Y2 = H2@Z
        R2 = R[idx_range,:]; R2 = R2[:,idx_range]  # this is a workaround; R2 = R[idx_range,idx_range] results in bugs
        #R2 = apply_gaussian_filter(R2, sigma)
        K2 = Z@Y2.T@LA.inv(Y2@Y2.T + R2)
        # Step 2: calculate tranformation matrix T
        A = np.identity(m) + Y2.T@LA.inv(R2)@Y2
        T = LA.inv(sqrtm(A))  
        # Step 3: update mean
        mean_temporary = ensemble_mean + K2@(y2 - ensemble_mean[idx_range])
        # Step 4: update Z
        Z_temporary = Z@T
        # Step 5: only keeping the n-th row of mean and Z
        ensemble_mean_new[idx] = mean_temporary[idx]
        Z_new[idx,:] = Z_temporary[idx,:]
        
        
    # Final step: after updating each grid,  Z -> ensemble
    ensemble_new = Z2member(Z_new,ensemble_mean_new)

    return ensemble_new, ensemble_mean_new
```

```python
# main loop: run the DA process
np.random.seed(10)
obs_err_std = np.sqrt(1);   
obs_err = np.random.randn(nature.shape[0], nature.shape[1]) * obs_err_std
obs = nature + obs_err
ensemble = []; ensemble_mean_history = []; ensemble_history = []; x_history=[]; x_mean_history = []
analysis_intv = int(10); 
m = 10 # ensemble size
n = 40  # L96 size
sigma = 3;
cutoff = 5     # round(sigma*3.4);    # localization cutoff scale

H = np.identity(n); R = np.identity(n) * obs_err_std**2;


import time
start_time = time.time()

initialization = True
## initialization
if initialization:
    for _ in range(m):
        initial_state = np.random.rand(n) # random number 
        for _ in range(1000):
            initial_state = lorenz96_step(initial_state)   # free run for 1000 steps
        ensemble.append(initial_state)  


## main loop
analysis_step = n_steps-6000
for ii in range((analysis_step-1) // analysis_intv):
    ensemble, ensemble_mean, ensemble_mean_history, x_history,x_mean_history, Z = lorenz96_forecast(ensemble,ensemble_mean_history,x_history,x_mean_history)
    y = obs[(ii+1)*analysis_intv,:]
    ensemble, ensemble_mean = lorenz96_analysis(ensemble, ensemble_mean, Z, H, y, R)
    
    ensemble_mean_history[-1] = ensemble_mean
    
    # inflation
    for jj in range(m):
        ensemble[jj] = ensemble_mean + (ensemble[jj]-ensemble_mean) * np.sqrt(1.03)  # optimization needed
    if ii % 100 == 0:
      print("analysis ", ii)
    else:
      print(".", end="")
print("Done")
        
ensemble_mean_history = np.array(ensemble_mean_history)
ensemble_history = np.array(ensemble_history)

end_time = time.time()
print("Elapsed time: ", end_time - start_time, "seconds")
```
